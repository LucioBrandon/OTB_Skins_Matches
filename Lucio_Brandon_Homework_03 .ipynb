**Q2** (1 point). Bin Continuous Values (`Age`) in `df` into Discrete Intervals (create a new column in `df` called `Age_group`) : `["20-", "20-40", "40-60", "60-80","80+" ]`  
*hint: you can choose to use `pd.cut(column, bins)` method*

```python
#With defined age group levels of Age_level as below
Age_bin = [i for i in range(int(min(df.Age)), 120, 20)]
print(Age_bin)

# you are not required to use levels defined in 'Age_level` create a new column as of "Age_grp".
# provide your answer here
df['Age_grp'] = pd.cut(df['Age'], Age_bin, labels = ["(0-20]", "(20-40]", "(40-60]", "(60-80]",
                                                     "(80+]" ] )

df.Age_grp.unique()
```
